## macierz rzadka

```python
import numpy as np
from scipy import sparse
eye = np.eye(10)
sm = sparse.csr_matrix(eye)
print( "arr NumPy:\n{}".format(eye))
print("\nmatrix SciPy in CSR: {}".format(sm))
```

## matplotlib nice intro!

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10,10,100)
y = np.sin(x)
plt.plot(x,y, marker="x")
plt.show()
```
